## Voting classifiers

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs',C=50 ,random_state=42, max_iter=5000)
poly_kernel_svm_clf = SVC(kernel="poly", degree=3, coef0=1, C=50, gamma='auto', probability= True) 
forest_clf = RandomForestClassifier(n_estimators=300, max_depth=30, n_jobs=-1, random_state=42 )

voting_clf = VotingClassifier(
    estimators=[('svc', poly_kernel_svm_clf) ,('lg', log_reg),('fr',forest_clf)], voting= 'soft', n_jobs=-1)
voting_clf.fit(train_data_scale, train_target)
```
